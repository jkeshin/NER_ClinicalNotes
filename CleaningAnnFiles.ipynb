```python
import pandas as pd
import os

training_data_dir = "\\track2-training_data_2"
training_dir = os.getcwd() + training_data_dir
training_dir
os.listdir(training_dir)
```

```python
training_data = []
tags = ["Reason", "Drug", "ADE"]
training_tags = {}
for filename in os.listdir(training_dir):
    if (".ann" in filename):
        with open(os.path.join(training_dir, filename), 'r') as f: # open in readonly mode
            doc_tags = []
            lines = f.readlines()
            for line in lines:
                if any(word in line for word in tags) and line[0] == 'T':
                    doc_tags.append(" ".join(line.split()))
        training_tags[filename] = doc_tags
                    
```
